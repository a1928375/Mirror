```python
def bonus_template(quadrant):
    "Make a board from the upper-left quadrant."
                                                        
    return mirror(list(map(mirror, quadrant.split())))     # split()用""分開所有的sections => 因為triple quotation marks中的
                                                           # sections是用""來分行
                                                           #且因為quadrant是string => split()後return的是list，如Fig. 1
                                                           #我們需要list代入mirror()，如Fig. 2
                                                           # map裡的mirror是先做左右鏡射，外面的mirror是做上下鏡射

def mirror(sequence): 
    
    return sequence + sequence[-2::-1]                     #從倒數第二個element開始，因為最後一個element是對稱中心 (中心不能重覆)，
                                                           #然後倒著走

print (mirror("|.....*"))
print (mirror("^._"))
print (bonus_template("""||||
                         |3.3
                         |.:.
                         |3.*
                         """))

print ("--------------------------------------------------------------------------------------------------------------------------")

SCRABBLE = bonus_template("""
|||||||||
|3..:...3
|.2...;..
|..2...:.
|:..2...:
|....2...
|.;...;..
|..:...:.
|3..:...*
""")

WWF = bonus_template("""
|||||||||
|...3..;.
|..:..2..
|.:..:...
|3..;...2
|..:...:.
|.2...3..
|;...:...
|...:...*
""")

print (SCRABBLE)                                  #鏡射完的result

print ("--------------------------------------------------------------------------------------------------------------------------")

print (WWF)                                       #鏡射完的result

print ("--------------------------------------------------------------------------------------------------------------------------")

def show(board):
    "Print the board."
    
    for i in range(len(board)):
        
        for j in range(len(board[i])):
            
            print (board[i][j], end= "")          #依序print出elements，不加空格
                
        print ()                                  #每print完一行 (每一組i底下的所有j)，就換行

# or        
        
#def show2(x):
    
#    for i in x:
        
#        for j in i:
            
#            print(j, end="")
        
#        print ()

print (show(SCRABBLE))                            # print出board

print ("------------------")

print (show(WWF))                                 # print出board
```
